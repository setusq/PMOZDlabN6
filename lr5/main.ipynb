1.4. Заполение пропусков данных медианным значением:

```python
median_total_income = df['total_income'].median()
df['total_income'] = df['total_income'].fillna(median_total_income)
df['total_income'].isnull().sum()
```

Почему в 'total_income' заполнить пропуски медианным значением - лучшее решение:
1. Сохранение центральной тенденции: Медиана является мерой центральной тенденции и не чувствительна к выбросам, что делает ее стабильной оценкой для дохода. Это может помочь избежать искажений в данных, вызванных крайне высокими или низкими значениями.
2. Сохранение структуры данных: Заполнение медианой позволяет сохранить общую структуру распределения доходов и избежать влияния экстремальных значений на статистику.
3. Меньшая чувствительность к выбросам: В сравнении с средним значением, медиана более устойчива к выбросам, что может быть важно, если в данных присутствуют необычные или ошибочные значения.
4. Простота и интерпретируемость: Заполнение медианой является относительно простым методом, легко интерпретируемым и реализуемым.

2.1 Обработка значений с аномалиями


```python
def delAnomaly(value): 
    if value < 0: value *= -1
    return value

df['days_employed'] = df['days_employed'].apply(delAnomaly)
df
```

Возможные причины таких аномалий:

1. Ошибки при сборе данных: В процессе сбора данных могли возникнуть технические ошибки или ошибки при вводе, в результате чего значения стажа были записаны с отрицательным знаком.
2. Проблемы с хранением данных: Ошибки при хранении данных могли привести к некорректному представлению значения стажа.
3. Проблемы с выгрузкой данных: При выгрузке данных из других источников могли возникнуть проблемы, в результате которых значения стажа были записаны с отрицательным знаком.
4. Ошибки при обработке данных: В ходе обработки данных могли возникнуть ошибки, в результате которых значения стажа были искажены.

Обработка отрицательных значений – это один из способов устранения аномалий, но также важно понимать источник проблемы и предпринимать меры для ее предотвращения в будущем.

2.2 Заполнение пропусков в 'days_employed' медианнымы значениями по стобцу

```python
median_days_employed = df['days_employed'].median()
df['days_employed'] = df['days_employed'].fillna(median_days_employed)
df['days_employed'].isnull().sum()
```

3. Замена вещественного типа данных в 'total_income'

```python
df['total_income'] = df['total_income'].astype(int)
df['total_income']
```

4. Удаление строк-дубликатов. Обработка неявных дубикатов.

```python
for column, series in df.items():
    if pd.api.types.is_string_dtype(series):
        if column == 'gender' or column == 'income_type' or column == 'purpose':
            continue
        df[column] = df[column].str.lower()
        df[column] = df[column].drop_duplicates()

df
```

4.1. Как был выбран метод:
Использование метода приведения строк к нижнему регистру (str.lower()) и удаление дубликатов (drop_duplicates()) может быть обоснованным в определенных сценариях:
1. Единообразие данных: Приведение всех строковых данных к нижнему регистру обеспечивает единообразие, что полезно при анализе данных. Это позволяет избежать несогласованности в данных, вызванной разным использованием регистра.
2. Устранение дубликатов: drop_duplicates() удаляет повторяющиеся значения, что может быть важно, если в данных есть дубликаты, которые могут привести к искажению результатов анализа.
3. Экономия памяти: Удаление дубликатов может также помочь экономии памяти, особенно если дубликаты занимают большой объем данных.
4. Улучшение производительности: Работа с данными в нижнем регистре может улучшить производительность при сравнении строк и выполнении операций сравнения.

4.2. Дубликаты в данных могут возникнуть по разным причинам:
1. Ошибка при вводе данных: При ручном вводе данных могут возникнуть опечатки или ошибки, приводящие к появлению дубликатов.
2. Проблемы при сборе данных: При сборе данных из разных источников могут возникнуть проблемы, такие как слияние данных из нескольких источников или технические сбои, приводящие к дубликатам.
3. Обновление данных: При обновлении данных новые записи могут быть внесены повторно, создавая дубликаты в наборе данных.
4. Проблемы с процессом выгрузки данных: В ходе процесса выгрузки данных из базы данных или других систем могут возникнуть ошибки, в результате которых данные могут быть дублированы.

5. Cоздаем новые датафреймы и удаляем старые

```python
df_education = df[['education']].drop_duplicates().reset_index(drop=True)
df_education['education_id'] = df_education.index

df_family_status = df[['family_status']].drop_duplicates().reset_index(drop=True)
df_family_status['family_status_id'] = df_family_status.index

df = df.merge(df_education)
df = df.merge(df_family_status)
df = df.drop('education', axis='columns')
df = df.drop('family_status', axis='columns')
df
```

6. Создааем столбец 'total_income_category' с категориями:

```python
df['total_income_category'] = df['total_income']

for column, series in df.items():
    if column == 'total_income_category':
        df[column] = df[column].astype(int)
        for i in range(len(series)):
            if series[i] <= 30000:
                df.at[i, column] = 'E'
            elif series[i] <= 50000:
                df.at[i, column] = 'D'
            elif series[i] <= 200000:
                df.at[i, column] = 'C'
            elif series[i] <= 1000000:
                df.at[i, column] = 'B'
            else:
                df.at[i, column] = 'A'
df
```

```python
df['purpose_category'] = df['purpose']

for column, series in df.items():
    if column == 'purpose_category':
        for i in range(len(series)):
            if series[i].__contains__('жиль') or series[i].__contains__('недвиж'):
                df.at[i, column] = 'операции с недвижимостью'
            elif series[i].__contains__('авто'):
                df.at[i, column] = 'операции с автомобилем'
            elif series[i].__contains__('свадь'):
                df.at[i, column] = 'проведение свадьбы'
            elif series[i].__contains__('образ'):
                df.at[i, column] = 'получение образования'
df
```

<b>Шаг 3.</b> Ответить на вопросы:

<li><b>Влияет ли количество детей на своевременное погашение кредита?</b></li> 

Да, так как расходы на детей могут влиять на финансовую стабильность, заемщики с детьми могут сталкиваться с увеличенными финансовыми обязательствами, что влияет на вероятность возврата кредита в срок.

<li><b>Как семейное положение влияет на возврат кредита?</b></li> 

Например, семейное положение, такое как брак или зарегистрированное партнерство, обычно связано с более низким уровнем просрочек. Это может объясняться более стабильными семейными обязательствами и ответственностью перед семьей.

<li><b>Как уровень дохода влияет на возврат кредита в срок?</b></li> 

Уровень дохода оказывает влияние на платежеспособность. Люди с более высоким доходом часто имеют более устойчивую финансовую ситуацию, что снижает вероятность просрочек, в то время как заемщики с низким доходом могут столкнуться с трудностями при возврате кредита.

<li><b>Как разные цели кредита влияют на своевременное погашение?</b></li> 

Исследования показывают, что заемщики, использующие кредит для инвестиций или образования, чаще выполняют свои финансовые обязательства. С другой стороны, кредиты на потребительские нужды могут иметь более высокий уровень просрочек из-за менее конкретных целей и финансовой неопределенности.
